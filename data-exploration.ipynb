```python
import math
import time
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# ---------- 1) Load tickers ----------
def load_nyse_tickers():
    return pd.read_csv("data/nyse-all-large-cap.csv")

# ---------- 2) Feature extraction ----------
def _safe(v):
    return None if v in (None, "None") else v

def compute_price_volume_features(ticker, trading_days=252):
    """
    Extracts:
      - avg_daily_return_12m : mean daily return over 1 year
      - vol_12m : annualized volatility
      - avg_daily_volume_12m : mean daily trading volume over 1 year
    """
    out = {"avg_daily_return_12m": np.nan, "vol_12m": np.nan, "avg_daily_volume_12m": np.nan}
    try:
        tk = yf.Ticker(ticker)
        d = tk.history(period="1y", interval="1d", auto_adjust=False)
        if len(d) >= 2:
            rets = d["Close"].pct_change().dropna()
            if len(rets) >= min(trading_days // 2, 60):
                tail = rets.tail(trading_days)
                out["avg_daily_return_12m"] = float(tail.mean())
                out["vol_12m"] = float(tail.std() * math.sqrt(252))
            out["avg_daily_volume_12m"] = float(d["Volume"].tail(trading_days).mean())
    except Exception:
        pass
    return out

def get_fundamentals(ticker):
    out = {"shares_out": np.nan, "pe_trailing": np.nan, "market_cap": np.nan}
    try:
        tk = yf.Ticker(ticker)
        fi = getattr(tk, "fast_info", None)
        mc = getattr(fi, "market_cap", None) if fi else None
        so = getattr(fi, "shares", None) if fi else None

        pe = None
        try:
            info = tk.info
            pe = _safe(info.get("trailingPE"))
            if mc is None:
                mc = _safe(info.get("marketCap"))
            if so is None:
                so = _safe(info.get("sharesOutstanding"))
        except Exception:
            pass

        out["market_cap"] = float(mc) if _safe(mc) is not None else np.nan
        out["shares_out"] = float(so) if _safe(so) is not None else np.nan
        out["pe_trailing"] = float(pe) if _safe(pe) is not None else np.nan
    except Exception:
        pass
    return out

def build_feature_table(tickers, throttle_every=8, pause=0.3):
    rows = []
    for i, t in enumerate(tickers["Symbol"], 1):
        r = compute_price_volume_features(t)
        f = get_fundamentals(t)
        rows.append({
            "ticker": t,
            "avg_daily_return_12m": r["avg_daily_return_12m"],
            "vol_12m": r["vol_12m"],
            "avg_daily_volume_12m": r["avg_daily_volume_12m"],
            "market_cap": f["market_cap"],
            "pe_trailing": f["pe_trailing"],
            "shares_out": f["shares_out"],
        })
        if i % throttle_every == 0:
            time.sleep(pause)
    return pd.DataFrame(rows)

# ---------- 3) Cleaning ----------
def clean_and_impute(df):
    d = df.copy()
    cols = [
        "avg_daily_return_12m",
        "vol_12m",
        "avg_daily_volume_12m",
        "market_cap",
        "pe_trailing",
        "shares_out",
    ]
    d[cols] = d[cols].replace([np.inf, -np.inf], np.nan)
    for c in cols:
        med = d[c].median(skipna=True)
        d[c] = d[c].fillna(med)
    return d

# ---------- 4) Clustering ----------
def cluster_and_select(df, k=5, random_state=42):
    cols = [
        "avg_daily_return_12m",
        "vol_12m",
        "avg_daily_volume_12m",
        "market_cap",
        "pe_trailing",
        "shares_out",
    ]
    X = df[cols].astype(float).values
    Z = StandardScaler().fit_transform(X)
    Zdf = pd.DataFrame(Z, index=df.index, columns=[f"Z_{c}" for c in cols])

    # Composite: emphasize high returns, high volume/liquidity, and size,
    # penalize volatility and valuation
    comp = (
        0.35 * Zdf["Z_avg_daily_return_12m"]
        - 0.20 * Zdf["Z_vol_12m"]
        + 0.15 * Zdf["Z_avg_daily_volume_12m"]
        + 0.15 * Zdf["Z_market_cap"]
        + 0.10 * Zdf["Z_shares_out"]
        - 0.10 * Zdf["Z_pe_trailing"]
    )

    work = df.copy()
    work["composite"] = comp

    km = KMeans(n_clusters=k, n_init=10, random_state=random_state)
    labels = km.fit_predict(Z)
    work["cluster"] = labels

    best_cluster = int(
        work.groupby("cluster")["composite"].mean().sort_values(ascending=False).index[0]
    )
    winners = (
        work[work["cluster"] == best_cluster]
        .sort_values("composite", ascending=False)
        .head(5)
        .reset_index(drop=True)
    )
    return winners, work


# ---------- 5) Main ----------
def main():
    print("Loading NYSE tickers ...")
    tickers = load_nyse_tickers()
    print("Loaded NYSE tickers ...")

    print("Building features via yfinance ...")
    feats = build_feature_table(tickers)
    feats = clean_and_impute(feats)

    # Save ALL features (including small/mid caps)
    feats.to_csv("data/nyse_large_cap_all_features.csv", index=False)
    print("Saved: nyse_all_features.csv (all tickers)")

    winners, work = cluster_and_select(feats)
    winners.to_csv("data/nyse_largecap_top5.csv", index=False)
    print("Saved: nyse_largecap_top5.csv")

    print("\nTop 5 clustered large-cap NYSE stocks:")
    print(winners)
```

```python
main()
```
