```python
import pandas as pd
```

```python
spy=pd.read_csv('data/etf/spy_20250102.csv', parse_dates=["Date-Time"],date_format="%Y-%m-%dT%H:%M:%S.%f%z",engine="python", sep = ',')
```

```python
spy.head()
```

```python
spy.shape
```

```python
# Convert to America/New_York timezone (if not already)
spy["Date-Time"] = spy["Date-Time"].dt.tz_convert("America/New_York")
```

```python
# Filter only regular market hours (09:30–16:00 Eastern)
df_market = spy[
    (spy["Date-Time"].dt.time >= pd.to_datetime("09:30:30").time()) &
    (spy["Date-Time"].dt.time <= pd.to_datetime("15:59:00").time())
]
```

```python
df_market.shape
```

```python
df_market.head(1000)
```

```python
base_cols = ["#RIC", "Date-Time"]
#base_cols = ["#RIC"]
# Dynamically build list for L1–L10
levels = []
for i in range(1, 11):  # L1–L10 inclusive
    levels += [
        f"L{i}-BidPrice", f"L{i}-BidSize",
        f"L{i}-AskPrice", f"L{i}-AskSize"
    ]
```

```python
spy_l10 = df_market[base_cols + levels]
```

```python
spy_l10
```

```python
# Total memory usage in bytes → convert to MB
mem_mb = spy_l10.memory_usage(deep=True).sum() / (1024 ** 2)
print(f"DataFrame size: {mem_mb:.2f} MB")
```

```python
spy_l10.head(5)
```

```python
# Keep the last row per (#RIC, millisecond)
spy_l10["ms_timestamp"] = spy_l10["Date-Time"].dt.floor("ms")
spy_l10_dedup = (
    spy_l10.drop_duplicates(subset=["#RIC", "ms_timestamp"], keep="last")
      .drop(columns="ms_timestamp")
)
```

```python
spy_l10_dedup.shape
```

```python
# Total memory usage in bytes → convert to MB
mem_mb = spy_l10_dedup.drop(columns="Date-Time").memory_usage(deep=True).sum() / (1024 ** 2)
print(f"DataFrame size: {mem_mb:.2f} MB")
```

```python
gld=pd.read_csv('data/etf/gld_20250102.csv', parse_dates=["Date-Time"],date_format="%Y-%m-%dT%H:%M:%S.%f%z",engine="python", sep = ',')
```

```python
gld.shape
```

```python
# Convert to America/New_York timezone (if not already)
gld["Date-Time"] = gld["Date-Time"].dt.tz_convert("America/New_York")
# Filter only regular market hours (09:30–16:00 Eastern)
gld_df_market = gld[
    (gld["Date-Time"].dt.time >= pd.to_datetime("09:30:30").time()) &
    (gld["Date-Time"].dt.time <= pd.to_datetime("15:59:00").time())]

base_cols = ["#RIC", "Date-Time"]
#base_cols = ["#RIC"]
# Dynamically build list for L1–L10
levels = []
for i in range(1, 11):  # L1–L10 inclusive
    levels += [
        f"L{i}-BidPrice", f"L{i}-BidSize",
        f"L{i}-AskPrice", f"L{i}-AskSize"
    ]
gld_l10 = gld_df_market[base_cols + levels]
gld_l10["ms_timestamp"] = gld_l10["Date-Time"].dt.floor("ms")
gld_l10_dedup = (
    gld_l10.drop_duplicates(subset=["#RIC", "ms_timestamp"], keep="last")
      .drop(columns="ms_timestamp")
)
```
